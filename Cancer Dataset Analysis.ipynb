```python
# Importing requisite libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from skimpy import skim
```

```python
# Reading the data file

cancer=pd.read_csv("cancerdata.csv")
```

```python
# Replacing the diagnosis values (char) with numeric values

# Malignant - 1 Benign - 0

cancer["diagnosis"]=cancer["diagnosis"].replace(["M","B"],[1,0])
```

```python
# Dropping the id column as it does not add any value

cancer=cancer.drop(["id"],axis=1)
```

```python
# Value count of our output variable (diagnosis)

cancer.diagnosis.value_counts()
```

```python
# Bar plot of value count of diagnosis

cancer["diagnosis"].value_counts().plot.bar()
label = ["B","M"]
plt.xticks(range(len(label)),label, rotation = 45)
plt.show()
```

```python
# min max scaler for standardisation of measurement scale of the variables

from sklearn.preprocessing import MinMaxScaler
min_max=MinMaxScaler()
min_max.fit_transform(cancer)
```

```python
# Converting the scaled output into a dataframe

cancer=pd.DataFrame(min_max.fit_transform(cancer))
```

```python
# Assigning the name of the columns to the dataframe

cancer.columns=['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst', 'compactness_wors',
       'concavity_worst', 'concave points_worst', 'symmetry_worst',
       'fractal_dimension_worst']
```

```python
# The min max of all the variables are 0 and 1

# Transposing the output for a clear view

cancer.describe().T
```

```python
# Correlation analysis

cancer_corr=cancer.corr(method="pearson")
cancer_corr
```

```python
range(len(cancer_corr))
```

```python
# Heatmap for correlation analysis

plt.figure(figsize=(25,20))
sns.heatmap(cancer_corr, annot= True, linewidths= 0.3 ,linecolor = "black", fmt = ".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Set a threshold for high correlation

threshold = 0.70

# Create a list to store the high correlation pairs

high_correlation_pairs = []

# Iterate through the correlation matrix and identify pairs with correlation above the threshold
# Use absolute value for correlation check

for i in range(len(cancer_corr)):
    for j in range(i + 1, len(cancer_corr)):
        if abs(cancer_corr.iloc[i, j]) > threshold:  
            high_correlation_pairs.append((cancer_corr.index[i], cancer_corr.index[j], cancer_corr.iloc[i, j]))

# Print the high correlation pairs

high_correlation_df = pd.DataFrame(high_correlation_pairs, columns=["Feature 1","Feature 2", "Correlation"])
high_correlation_df
```

# Logistic Regression

```python
# Storing the output and input variables separately

x=cancer.drop(["diagnosis"],axis=1)
y=cancer["diagnosis"]
```

```python
# Creating test and train datasets 

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30, random_state=42)
```

```python
# Shape of test and train datasets

x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
# Importing our logistic regression and training the model

from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression()
log_reg.fit(x_train,y_train)
```

```python
log_reg.intercept_
```

```python
log_reg.coef_
```

```python
# Predicting output of test data

prediction=log_reg.predict(x_test)
prediction
```

```python
# The probability of each test data point of having output 0 or 1

log_reg.predict_proba(x_test)
```

```python
# Creating ROC-AUC plot for our model

from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
logit_roc_auc = roc_auc_score(y_test,log_reg.predict(x_test))
fpr, tpr, thresholds = roc_curve(y_test,log_reg.predict_proba(x_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```

```python
# Model evaluation

from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,prediction))
```

```python
# Confusion matrix

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,prediction)
```

```python
# Heatmap of confusion matrix

sns.heatmap(confusion_matrix(y_test,prediction),annot=True)
```

```python
# Accuracy score of logistic regression model

from sklearn.metrics import accuracy_score
accuracy_score(y_test, prediction)
```

# XGBoost

```python
# Splitting the dataset into input and output

# In "x", we're storing only predictor variables; in "y" we're storing target variable

x1=cancer.drop(["diagnosis"],axis=1)

y1=cancer["diagnosis"]
```

```python
# Splitting the datasets into training and testing input and ouput variables

from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.3, random_state=42, stratify=y1)
```

```python
from xgboost import XGBClassifier
```

```python
# Creating model with parameters

clf=XGBClassifier(max_depth=8, learning_rate=0.05, n_estimators=1000, objective='binary:logistic', booster='gbtree', subsample=0.8, min_child_weight=1, colsample_bytree=0.8, tree_method="exact")
```

```python
# Creating the model on training data

XGB=clf.fit(x1_train,y1_train)
prediction_1=XGB.predict(x1_test)
```

```python
# Measuring accuracy on Testing Data

from sklearn import metrics
print(metrics.classification_report(y1_test, prediction_1))
```

```python
# Confusion matrix

from sklearn import metrics
conf_mat = metrics.confusion_matrix(y1_test, prediction_1)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = conf_mat, display_labels = ["0","1"])
cm_display.plot()
```

```python
# Accuracy of our predictions

from sklearn.metrics import accuracy_score
accuracy_score(y1_test, prediction_1)
```
